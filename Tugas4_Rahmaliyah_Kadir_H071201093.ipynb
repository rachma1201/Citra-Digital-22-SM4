#1. Segmentasi citra yang diberikan sebagai berikut:

```python
img = cv2.imread('cell darah.png',0)
plt.figure(figsize = (10,6))
plt.axis('off')
plt.imshow(img,'gray')
```

```python
# digunakan median blur untuk menghaluskan tepi objek pada citra
# ini diperlukan agar thresholding memberikan hasil lebih baik
imgg = cv2.medianBlur(img,5)
```

```python
# Lakukan Thresholding
# Binary Threshold
ret,th1 = cv2.threshold(imgg,165,255,cv2.THRESH_BINARY_INV)
# Plotting
plt.figure(figsize = (10,6))
plt.axis('off')
plt.imshow(th1,'gray')
```
