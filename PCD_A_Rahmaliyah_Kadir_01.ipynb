<h1>a. Transformasi menjadi citra biner mnenggunakan thresholding dengan nilai.</h1>

```python
T = 100
img_A = np.array(mountain_img)
img_B = np.zeros(mountain_img.shape)
for x in range(0, len(img_A)):
    for y in range(0, len(img_B[0])):
      if (any(img_A[x][y]<T)):
        img_B[x][y] = 0
      else:
        img_B[x][y] = 1
```

```python
plt.imshow(img_B, cmap = 'gray')
```

<h1>b. Transformasi menjadi citra negatif.</h1>

```python
img_A = np.array(mountain_img)
img_B = np.zeros(mountain_img.shape)
for x in range(0, len(img_A)):
    for y in range(0, len(img_B[0])):
      img_B[x][y] = 255 - img_A[x][y]
```

```python
plt.imshow(img_B.astype('uint8'), cmap = 'gray')
```

<h1>c. Tingkatkan kecerahan citra dengan nilai b = 100</h1>

```python
b = 100
amma_A = np.array(mountain_img)
amma_B = np.zeros(mountain_img.shape)
for x in range(0, len(amma_A)):
    for y in range(0, len(amma_B[0])):

      ams =  amma_A[x][y] + b
      if (any(ams<0)):
        amma_B[x][y] = 0
      elif (any(ams>255)):
        amma_B[x][y] = 255
      else:
        amma_B[x][y] = ams
```

```python
plt.imshow(amma_B.astype('uint8'), cmap = 'gray')
```

<h1>2. Dengan menggunakan citra girl-red.bmp konversi citra menggunakan nilai rata-rata RGB.</h1>

```python
girl_red = mping.imread('/content/drive/MyDrive/TGS_PCD_01/TGS01/girl-warna.bmp')
plt.imshow(girl_red)
```

```python
B = np.array(girl_red[:,:,0])
G = np.array(girl_red[:,:,1])
R = np.array(girl_red[:,:,2])
```

```python
rgb_rammA = np.zeros(girl_red.shape)

for x in range(0, len(girl_red)):
  for y in range(0, len(girl_red[0])):
    rgb_rammA[x][y] = ((R[x][y] + G[x][y] + B[x][y]) / 3)

plt.imshow(rgb_rammA.astype('uint8'))
```

<h1>3. Kerjakan soal nomor 2 dengan menggunakan rumus Y = 0.299R + 0.587G + 0.144B</h1>

```python
rgb_rammB = np.zeros(girl_red.shape)
```

```python
for x in range(0, len(girl_red)):
  for y in range(0, len(girl_red[0])):
    amm = (0.299*R[x][y] + 0.587*G[x][y] + 0.144*B[x][y])

    if (any(amm < 0)):
      rgb_rammB[x][y] = 0
    elif (any(amm > 255)):
      rgb_rammB[x][y] = 255
    else :
      rgb_rammB[x][y] = amm

plt.imshow(rgb_rammB.astype('uint8'))
```
